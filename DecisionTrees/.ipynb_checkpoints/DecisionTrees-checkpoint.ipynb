# Implementing a decision tree
The aim of the decision tree is to predict the job category based on other factors

## Entropy function

```python
#Calculating entorpy
def entropy(dataset, target):
    categories = dataset[target].unique()
    total_count = dataset[target].shape[0]
    e = 0
    for c in categories:
        p = (dataset[dataset[target] == c].shape[0]) / total_count
        e += -1 * p * (np.log2(p) if p != 0 else 0)
    return e
```

```python
df_test1 = df_lite.head()
df_test1
```

```python
entropy(df_test1, "Job_Category")
```

```python
#Calculate information gain
def informationGain(parent, children , target):
    parent_size = parent[target].shape[0]
    info_gain = entropy(parent, target)
    for child in children:
        p_child = child[target].shape[0] / parent_size
        info_gain -= p_child * entropy(child, target)
    return info_gain
```

```python
#Testing informationGain function
df_0 = df_test1[(df_test1['Job_Category'] == 0) | (df_test1['Job_Category'] == 1) ]
df_1 = df_test1[df_test1['Job_Category'] == 2 ]
df_2 = df_test1[df_test1['Job_Category'] == 3 ]

informationGain(df_test1, [df_0, df_1, df_2], "Job_Category")
```

```python
#Mean squared error loss from baseline=mean
def mse(y):
    if y.shape[0] == 0:
        return 0
    mean = np.mean(y)
    residuals_sq = (y - mean)**2
    return np.sum(residuals_sq) 
```

```python
mse(y_train[0:5])
```

```python
#return the INDICES of the (x,y) in the left node and the (x,y) in the right node (LEFT: True, RIGHT: FALSE)
def split_binary(x_targets):
    left_indices = []
    right_indices = []
    for i,x_target in enumerate(x_targets):
        if x_target == 0:
            #X is False
            right_indices.append(i)
        else:
            #X is True
            left_indices.append(i)
    return left_indices, right_indices, None
```

```python
def split_cont_choose(x_targets, loss_func):
    #calculate all available midpoints, taking it as a dividing line and seeing which one is the best
    left_indices = []
    right_indices = []
    original_size = x_targets.shape[0]
    #calculate midpoints
    #we calculate the midpoints by add an array to its shifted version and removing the last element
    #this is much faster since computations are vectorized on GPU
    x_targets_shifted = np.append( x_targets[1:], x_targets[0])
    midpoints = (x_targets + x_targets_shifted) * 0.5
    midpoints = midpoints[:-1]

    mse_midpoints = []
    #we now should decide what to split on 
    for midpoint in midpoints:
        left_indices = np.where(x_targets >= midpoint)
        right_indices = np.where(x_targets < midpoint)

        left_data = x_targets[left_indices]
        right_data = x_targets[right_indices]
        #calculate the weighted sum
        left_data_size = left_data.shape[0]
        right_data_size = right_data.shape[0]
        
        weighted_mse_sum = (left_data_size / original_size) * loss_func(left_data) + (right_data_size / original_size) * loss_func(right_data)
        mse_midpoints.append(weighted_mse_sum)

    min_mse_index = np.argmin(mse_midpoints, axis=0)

    #perform split
    left_indices = np.where(x_targets >= midpoints[min_mse_index])
    right_indices = np.where(x_targets < midpoints[min_mse_index])
    return left_indices, right_indices, midpoints[min_mse_index]

split_cont_choose(x_hot_train[0:5,11], mse)
```

```python
def split_cont(x_targets, midpoint):
    left_indices = []
    right_indices = []
    original_size = x_targets.shape[0]

    left_indices = np.where(x_targets >= midpoint)
    right_indices = np.where(x_targets < midpoint)
    return left_indices, right_indices
```

```python
#return the category to split on
def choose_split(x, y, is_cont ,loss_func):
    mse_categories = []
    original_size = x.shape[0]
    #loop over all categories and record their weighted mse, the min wins
    for c in range(x.shape[1]):
        
        li, ri, possible_midpoint = (split_binary(x[:,c]) if is_cont == 0 else split_cont_choose(x[:,c], mse))
        left_data = y[li]
        right_data = y[ri]

        left_data_size = left_data.shape[0]
        right_data_size = right_data.shape[0]
        
        weighted_mse_sum = (left_data_size / original_size) * loss_func(left_data) + (right_data_size / original_size) * loss_func(right_data)
        print(f"The Weighted Sum Of Category NO {c} :{weighted_mse_sum}")
        mse_categories.append(weighted_mse_sum)
    max_mse_index = np.argmin(mse_categories, axis=0)
    return max_mse_index,possible_midpoint
```

```python
is_cont = ((x_hot_train.shape[1] - 1) * [0]) + [1]
best_choice = choose_split(x_hot_train, y_hot_train, is_cont, mse)
```

```python
class DecisionTreeRegressor:
    def __init__(self, right, left, index, mid, value):
        self.right = right
        self.left = left
        self.index = index
        self.mid = mid
        self.value = value
    
def BuildDecisionTreeRegressorIMPL(x, y, is_cont, max_depth, dtr):
    #Searching for the best way to split

    if max_depth == 0:
        dtr.value = np.mean(y)
    category, possible_midpoint = choose_split(x,y, is_cont, mse)

    dtr.index = category
    dtr.mid = possible_midpoint
    dtr.value = np.mean(y)
    
    li, ri =([],[])
    if is_cont[category] == 1:
        assert possible_midpoint != None
        li, ri = split_cont(x, possible_midpoint)
    else:
        li, ri, _ = split_binary(x)

    x_left = x[li]
    y_left = y[li]        

    x_right = x[ri]
    y_right = y[ri]

    dtr.right = DecisionTreeRegressor(None, None, None, None, None)
    dtr.left = DecisionTreeRegressor(None, None, None, None, None)
    
    BuildDecisionTreeRegressorIMPL(x_right, y_right, is_cont, max_depth - 1, dtr.right)
    BuildDecisionTreeRegressorIMPL(x_left, y_left, is_cont, max_depth - 1, dtr.left)
```
